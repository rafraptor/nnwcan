```python
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm
```

```python
example_file = "https://upload.wikimedia.org/wikipedia/commons/thumb/1/15/WildRat.jpg/275px-WildRat.jpg"
image = imread(example_file, as_gray=True)
plt.imshow(image, cmap = cm.gray)
plt.show()
```

```python
print(type(image))
```

```python
print(image.shape)
```

```python
image_2 = image[50:150, 50:150]
plt.imshow(image_2, cmap = cm.gray)
plt.show()
```

```python
image_3 = resize(image_2, (30,30), mode = 'symmetric')
plt.imshow(image_3, cmap = cm.gray)
plt.show()
```

```python
from lxml import objectify
xml = objectify.parse(open("Source Code/XMLData.xml"))
root = xml.getroot()
df = pd.DataFrame(columns=('Number', 'String', 'Boolean'))
for i in range(4):
    obj = root.getchildren()[i].getchildren()
    row = dict(zip(['Number', 'String', 'Boolean'],
                  [obj[0].text, obj[1].text, obj[2].text]))
    row_s = pd.Series(row)
    row_s.name = i
    df = df.append(row_s)
```
